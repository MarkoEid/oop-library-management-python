# Implementing a Library Management System (OOP)
## Objective: Create classes to represent a basic library system.
## Requirements:
## Classes: Book, Library, and Member.
* ### Book: Has properties like title, author, ISBN, available_copies.
* ### Library: Manages a collection of books and has methods to add, remove, and search for books.
* ### Member: Can borrow and return books.

```python
class Book:
    def __init__(self,title,author,ISBN,available_copies):
        self.title=title
        self.author=author
        self.ISBN=ISBN
        self.available_copies=available_copies
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self,book):
        self.books.remove(book) 
        
    def search_book(self,title):
        for i in self.books:
            if title==i:
                print("the book is found")
            else:
                print("not found")
        
class Member:

    def __init__(self):
        self.books = []

    
    def borrow_book(self,title):
        self.books.append(title)
        
    def return_book(self,N_book):
        self.books.remove(N_book)


l1=Library()
l1.add_book("science")
l1.search_book("science")
```
